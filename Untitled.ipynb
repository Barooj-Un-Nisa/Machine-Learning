```python
import matplotlib.pyplot as plt
import numpy as np
import PIL as image_lib
import tensorflow as tf
from tensorflow.keras.layers import Dense,Flatten
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
import pathlib
```

```python
data_directory = r"C:\Users\hp\Documents\Projects\Machine-learning\TransferLearning\assignment_4\Data_set"
```

```python
img_height,img_width = 180,180
batch_size = 32
#this code creates the training subset of 80% data.
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
data_directory,
validation_split = 0.2,
subset="training",
seed=123,
image_size=(img_height,img_width),
batch_size=batch_size)
```

```python
validation_ds = tf.keras.preprocessing.image_dataset_from_directory(
data_directory,
validation_split = 0.2,
subset="validation",
seed=123,
image_size=(img_height,img_width),
batch_size=batch_size)
```

```python
#visualizing the images from the dataset
plt.figure(figsize=(10,10))
for images,labels in train_ds.take(1):
    for var in range(6):
        ax = plt.subplot(3, 3, var+1)
        plt.imshow(images[var].numpy().astype("uint8"))
        plt.axis("off")
```

```python
#importing pre-trained ml model
#ResNet-50 from keras library

demo_resnet_model = Sequential()
pretrained_model_for_demo=tf.keras.applications.ResNet50(input_shape = (180,180,3),include_top = False,weights = 'imagenet',
                                                        pooling='avg',classes = 2)
for each_layer in pretrained_model_for_demo.layers:
    each_layer.trainable = False
demo_resnet_model.add(pretrained_model_for_demo)
```

```python
#the following code specifies to add a fully connected output layer to the model where the learning can happen
demo_resnet_model.add(Flatten())

demo_resnet_model.add(Dense(512,activation='relu'))
demo_resnet_model.add(Dense(5,activation='softmax'))
```

```python
#model training
demo_resnet_model.compile(optimizer=Adam(learning_rate=0.001),loss='sparse_categorical_crossentropy',metrics=['accuracy'])
history = demo_resnet_model.fit(train_ds,validation_data=validation_ds,epochs=1
                               )
```

```python
#before running the model on the dataset, the images must go through preprocessing steps. These
#steps ensure that each image's dimensions correspond to what the model trained on

#run the prediction on the sample image of a disease class 1
import cv2
sample_image=cv2.imread(r"C:\Users\hp\Documents\Projects\Machine-learning\TransferLearning\assignment_4\Data_set\class 1\1_6\10253_idx5_x701_y651_class1.png")
sample_image_resized=cv2.resize(sample_image,(img_height,img_width))

sample_image=np.expand_dims(sample_image_resized,axis=0)

#command for prediction
image_pred=demo_resnet_model.predict(sample_image)
image_pred
class_names = ['class 0', 'class 1']
# command to produce a human readable output
image_output_class=class_names[np.argmax(image_pred)]
print("The predicted class is",image_output_class)
```
